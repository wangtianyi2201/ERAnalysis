# Send a Message to GPT and Receive a Response
Write code to send a message to the GPT model using the OpenAI API and receive the response.

```python
# Send a Message to GPT and Receive a Response
def send_message_to_gpt(message, model="gpt-3.5-turbo"):
    response = openai.ChatCompletion.create(
        model=model,
        messages=[{"role": "user", "content": message}]
    )
    return response.choices[0].message.content

# Example usage
user_message = "Hello, GPT!"
response = send_message_to_gpt(user_message)
print(response)
```

# Link Notebook Output to a Specific Chat Thread
Implement logic to associate messages with a specific chat thread ID, so that the conversation context is maintained.

```python
# Link Notebook Output to a Specific Chat Thread
chat_threads = {}

def send_message_with_thread(message, thread_id, model="gpt-3.5-turbo"):
    if thread_id not in chat_threads:
        chat_threads[thread_id] = []
    
    chat_threads[thread_id].append({"role": "user", "content": message})
    
    response = openai.ChatCompletion.create(
        model=model,
        messages=chat_threads[thread_id]
    )
    
    reply = response.choices[0].message.content
    chat_threads[thread_id].append({"role": "assistant", "content": reply})
    
    return reply

# Example usage
thread_id = "my_chat_thread"
user_message = "Tell me about machine learning."
response = send_message_with_thread(user_message, thread_id)
print(response)
```

# Display and Store Chat History
Display the ongoing chat history in the notebook and optionally store it for future reference.

```python
# Display and Store Chat History
def display_chat_history(thread_id):
    if thread_id in chat_threads:
        for msg in chat_threads[thread_id]:
            print(f"{msg['role'].capitalize()}: {msg['content']}")
    else:
        print("No chat history for this thread.")

def save_chat_history(thread_id, filename="chat_history.json"):
    if thread_id in chat_threads:
        with open(filename, 'w') as f:
            json.dump(chat_threads[thread_id], f)
        print(f"Chat history saved to {filename}")
    else:
        print("No chat history to save.")

# Example usage
display_chat_history(thread_id)
save_chat_history(thread_id)
```
